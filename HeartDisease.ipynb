```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/selected_columns.csv')

# Grouping the data by 'LocationAbbr' and calculate the mean 'Data_Value'
grouped_data = df.groupby('LocationAbbr')['Data_Value'].mean().reset_index()

# Creating a bar graph
plt.figure(figsize=(10, 6))
plt.bar(grouped_data['LocationAbbr'], grouped_data['Data_Value'])

# Adding  titles and labels
plt.title('Year 2014')
plt.xlabel('Location Abbreviation')
plt.ylabel('Average Data Value')

plt.xticks(rotation=100)
plt.tight_layout()
plt.savefig('Year 2014.png')
plt.show()

```
